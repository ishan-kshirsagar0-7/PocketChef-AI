```python
import requests
import base64
from IPython.display import Image, display

def get_recipe_from_api(ingredients):
    # Construct the URL
    base_url = "https://pocketchef-ai.onrender.com/get-recipe/"
    query_params = {'text': ingredients}
    response = requests.get(base_url, params=query_params)

    # Check if the request was successful
    if response.status_code == 200:
        return response.json()  # or `response.text` if response is plain text
    else:
        return f"Error: {response.status_code}"


ingreds = """
tomatoes, onions, garlic, bell peppers, carrots, potatoes, spinach, mushrooms, ginger, cilantro, basil, thyme, pasta
"""
data = get_recipe_from_api(ingreds)
print("Recipe: \n")
print(data["recipe"]["recipe_name"])
print("\nInstructions: \n")
print(data["recipe"]["instructions"])
img = base64.b64decode(data["photo"])
display(Image(data=img))
urls = data["links"]
print(f"\nYouTube Links: \n{urls}")
```

```python
import requests
import base64
from IPython.display import Image, display

def get_recipe_from_image(image_path):
    # Define the API endpoint
    api_url = 'https://pocketchef-ai.onrender.com/post-recipe/'
    files = {'file':open(image_path,'rb')}
    res = requests.post(url=api_url,files=files)
    # print(res)
    final = res.json()
    return final

# Example usage:
result = get_recipe_from_image('pbm.jpg')
# print(result)
print("Recipe: \n")
print(result["recipe"]["recipe_name"])
print("\nInstructions: \n")
ins = result["recipe"]["instructions"]
for i in ins:
  print(i)
img2 = base64.b64decode(result["photo"])
display(Image(data=img2))
urls2 = result["links"]
print(f"\nYouTube Links: \n{urls2}")
```
